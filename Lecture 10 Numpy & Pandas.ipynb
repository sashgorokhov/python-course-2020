#### Memory view
Слайс numpy не создает копию данных (в отличие от списков python), а возвращает так называемый view.
Это значит, что если поменять значение в слайсе, то значение в оригинальном массиве также изменится

````python
a_slice = A[0:2, 0:2]
a_slice[0, 0] = -22
A
````

### pd.Series
Проиндексированный одномерный массив - сопоставляет каждому значению массива уникальный ключ

Представим, мы работаем с некоторыми данными о результатах тестирования лекарства для нормализации давления. <br/>
В качестве индекса будем использовать "имена" пациентов. <br/>
В массиве хранится информация о том, эффективно ли лекарство для конкретного пациента.

````python
patients = ["a", "b", "c", "d", "e", "f"]
effectiveness = [True, True, True, True, False, False]

effective_series = pd.Series(effectiveness, index=patients)
effective_series
````

Если нужно больше информации, одного массива уже недостаточно. Такие данные удобно хранить в табличной форме.

### pd.DataFrame
Проиндексировання таблица. Можно рассматривать DataFrame как словарь элементов Series

В роли индекса - снова имена пациентов. <br/>
В каждой колонке хранится значение отдельного признака для всех пациентов. <br/>
Каждая строка - набор всех признаков для конкретного пациента. 

````python
columns = {
    'gender': ['F', 'M', 'M', 'M', 'F', 'M'],
    'sys_initial': [120, 126, 130, 115, 150, 117],
    'dia_initial': [75, 85, 90, 87, 90, 74],
    'sys_final': [115, 123, 130, 118, 130, 121],
    'dia_final': [70, 82, 92, 87, 85, 74],
    'drug_admst': effectiveness
}

df = pd.DataFrame(columns, index=patients)
````

### groupBy
Возвращает итератор, каждый элемент которого - фрейм с соответствующим значением

````python
type(df.groupby('drug_admst'))
````

````python
for value, group in df.groupby('drug_admst'):
    print("Value: {}".format(value))
    print("Group DataFrame:")
    print(group)
````

````python
df.groupby('drug_admst').agg({
    'sys_initial': ['min', 'max'], 
    'sys_final': np.mean
})
````

### Joining

````python
hospitals = pd.DataFrame({
        "name" : ["City 1", "City 2", "City 3"],
        "address" : ["Address 1", "Address 2", "Address 3"],
        "city": ["City 1", "City 2", "City 3"] 
    }, index=["H1", "H2", "H3"])

hospitals
````

````python
df['hospital_id'] = ['H1', 'H2', 'H2', 'H3', 'H3', 'H3']
df
````

````python
full_df = df.join(hospitals, on='hospital_id')
full_df
````
